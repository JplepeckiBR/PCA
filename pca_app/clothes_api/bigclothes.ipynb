```python
import pandas as pd

# Load the dataset
file_path = '/Users/nattybatty/Downloads/Data - Copy.csv'
df = pd.read_csv(file_path)


# Filter out rows where the 'images' column is missing or empty
df_filtered = df.dropna(subset=['img'])


# Select only the relevant columns: 'name', 'color', and 'images'
df_filtered = df_filtered[['title', 'img']]

# Display the resulting DataFrame
df_filtered
```

```python
import requests
from PIL import Image
from io import BytesIO
import pandas as pd

# Function to convert RGB to hex
def rgb_to_hex(rgb):
    return '#{:02x}{:02x}{:02x}'.format(*rgb)

# Function to get the dominant color from an image URL using PIL
def get_dominant_color(image_url):
    try:
        response = requests.get(image_url)
        response.raise_for_status()  # Check for HTTP errors
        img = Image.open(BytesIO(response.content))
        img = img.convert('RGB')  # Ensure image is in RGB mode
        img = img.resize((1, 1))  # Resize image to 1x1 pixel
        dominant_color = img.getpixel((0, 0))  # Get the color of that pixel
        return rgb_to_hex(dominant_color)
    except requests.exceptions.RequestException as e:
        print(f"Request error for image {image_url}: {e}")
    except IOError as e:
        print(f"IO error for image {image_url}: {e}")
    except Exception as e:
        print(f"Unexpected error for image {image_url}: {e}")
    return None

# Apply the dominant color extraction to the images column
df_filtered['dominant_color'] = df_filtered['img'].apply(get_dominant_color)

# Drop rows where the color could not be extracted
df_cleaned = df_filtered.dropna(subset=['dominant_color'])

# Select only the relevant columns
df_final = df_cleaned[['title', 'dominant_color']]

# Display the resulting DataFrame
df_final
```
